## Exercise 3

Repeat the previous exercise with $h(x)=\frac{x+9}{2x-1}$, still considering what happens as $x$ approaches 0.5 from each side.

```python
function_e3 = (x + 9) / (2*x - 1) #h(x)
function_e3
```

```python
#Evaluating h(x) as x approaches 0.5 from the left

print(function_e3.subs(x, 0.49))
print(function_e3.subs(x, 0.499))
print(function_e3.subs(x, 0.4999))
print(function_e3.subs(x, 0.49999))
print(function_e3.subs(x, 0.499999))
```

As the value of x approaches 0.5 from the left, the function approaches a smaller value and as the value of x approaches 0.5 from the right, the function approaches a higher value. Therefore there is no convergence of the function h(x) at a single point and this is an example of an Unbounded Limit.

## Exercise 5

In the slides for today, there were five examples of ways a limit could diverge (Divergent Example 1 through Divergent Example 5).  For each one, evaluate the corresponding limit in SymPy and by doing so, explain what each piece of output means.  Does SymPy ever make a mistake?

```python
#Divergent Example 1
function_1 = 1 / (sin(x)*sin(x))
limit(function_1, x, 0 , '-')
```

The limit of the function is infinity which means that as x approaches 0 from the left, the function approaches positive infinity.

```python
#Divergent Example 2
function_2 = (3*x - 9) / (2*x - 2)
limit(function_2, x, 1, '+')
```

The limit of the function is negative infinity which means that as x approaches 1 from the right, the function approaches negative infinity.

```python
#Divergent Example 3
function_3 = (ln(x + 10)) / (x + 3)
print (limit(function_3, x, -3, '-'))
print (limit(function_3, x, -3, '+'))
```

The limit of the function as x approaches -3 from the left is negative infinity and the limit of the function as x approaches -3 from the right is positive infinity.

```python
#Divergent Example 4
function_4 = x / abs(3*x)
print (limit(function_4, x, 0, '-'))
print (limit(function_4, x, 0, '+'))
```

The limit of the function as x approaches 0 from the left is -1/3 which means that the y value is -1/3 for every value of x that is less than 0. 
The limit of the function as x approaches 0 from the right is 1/3 which means that the y value is 1/3 for every value of x that is greater than 0.

```python
#Divergent Example 5
function_5 = cos(1/x)
limit(function_5, x, 0, '+')
```

The above limit means that as x approaches 0 from the right, the value of the function is limited to the bounds of -1 and 1. This is an example of Sympy making a mistake.
